# INT8 / INT4 QAT for ResNet-18 (NVIDIA ModelOpt)

Fine-tunes a pre-trained FP32 ResNet-18 with ModelOpt fake-quant.
Optionally applies **input pixel quantization** (1/2/4/8 bits) before normalization.

Per-config checkpoints land under `checkpoints/qat/<PRECISION>_in<N>b/` so multiple
runs (different `INPUT_QUANT_BITS`) can coexist without overwriting each other.

**Headless run** (e.g. inside the docker image):

```bash
jupyter nbconvert --to notebook --execute training/train_qat_modelopt.ipynb \
  --output train_qat_modelopt.executed.ipynb
```


## Setup

````python
import os
import random
import sys
from pathlib import Path

ROOT = Path.cwd().resolve()
if ROOT.name == "training":
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))
sys.path.insert(0, str(ROOT / "src" / "qat_modelopt"))

import numpy as np
import PIL.ImageFile
PIL.ImageFile.LOAD_TRUNCATED_IMAGES = True
import torch
import torch.nn as nn
from torch.amp.grad_scaler import GradScaler
from torch.utils.data import DataLoader
from torchvision import transforms

from quantize import get_quant_cfg, get_model, quantize_model, restore_modelopt_state
from train_utils import train_one_epoch, validate, save_checkpoint, load_training_state
from data import build_train_holdout_split, Quantize01
````

## Configuration

Edit the values below. `INPUT_QUANT_BITS` quantizes pixel values in `[0, 1]`
to `2^N` levels **before** ImageNet normalization. Use `None` or `0` to skip.
Valid: `1, 2, 4, 8`.

`OUTPUT_DIR` is derived from `PRECISION` and `INPUT_QUANT_BITS` so each config
saves into its own subdirectory.


````python
DATA              = "/home/pf4636/imagenet"
CHECKPOINT        = str(ROOT / "checkpoints" / "best.pth")

PRECISION         = "int8"        # "int8" or "int4"
INPUT_QUANT_BITS  = None          # None / 0 -> off; 1, 2, 4, or 8 -> quantize input

EPOCHS            = 15
BATCH_SIZE        = 64
NUM_WORKERS       = 8
LR                = 1e-4
MOMENTUM          = 0.9
WEIGHT_DECAY      = 1e-4
NUM_CALIB_BATCHES = 8
NUM_CLASSES       = 100
SEED              = 42
GPU               = None          # int gpu index, or None for default

# Derived: each (precision, input-quant-bits) combo gets its own dir.
_RUN_TAG          = f"{PRECISION}_in{INPUT_QUANT_BITS or 0}b"
OUTPUT_DIR        = str(ROOT / "checkpoints" / "qat" / _RUN_TAG)

RESUME            = None          # path to a QAT epoch .pth, or None for fresh run
RESUME_MOSTATE    = None          # paired _mostate.pt; inferred from RESUME if None

print(f"[Config] tag={_RUN_TAG}  output_dir={OUTPUT_DIR}")
````

## Seed and device

````python
def set_seed(seed: int) -> None:
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(SEED)

if GPU is not None:
    torch.cuda.set_device(GPU)
device = torch.device(
    f"cuda:{GPU}" if GPU is not None
    else "cuda" if torch.cuda.is_available()
    else "cpu"
)
print(f"[Device] {device}")
````

## Data loaders

Calibration draws from `train_subset` to avoid leaking holdout statistics.


````python
_IMAGENET_MEAN = [0.485, 0.456, 0.406]
_IMAGENET_STD  = [0.229, 0.224, 0.225]

def _maybe_quantize(num_bits):
    if num_bits is None or int(num_bits) == 0:
        return []
    return [Quantize01(int(num_bits))]

normalize = transforms.Normalize(mean=_IMAGENET_MEAN, std=_IMAGENET_STD)
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4),
    transforms.ToTensor(),
    *_maybe_quantize(INPUT_QUANT_BITS),
    normalize,
])
val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    *_maybe_quantize(INPUT_QUANT_BITS),
    normalize,
])

train_subset, holdout_subset = build_train_holdout_split(
    data_root=DATA,
    num_classes=NUM_CLASSES,
    seed=SEED,
    train_transform=train_transform,
    eval_transform=val_transform,
)

train_loader = DataLoader(
    train_subset, batch_size=BATCH_SIZE, shuffle=True,
    num_workers=NUM_WORKERS, pin_memory=True, drop_last=True,
)
val_loader = DataLoader(
    holdout_subset, batch_size=BATCH_SIZE, shuffle=False,
    num_workers=NUM_WORKERS, pin_memory=True,
)
calib_loader = DataLoader(
    train_subset, batch_size=BATCH_SIZE, shuffle=True,
    num_workers=0, pin_memory=True,
)

print(f"[Input quant] bits = {INPUT_QUANT_BITS or 'off'}")
````

## Model

````python
model = get_model(CHECKPOINT, num_classes=NUM_CLASSES).to(device)
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
````

## Quantization graph

Fresh run: validate the FP32 baseline, then PTQ-calibrate via ModelOpt.
Resume: restore the saved quantization graph (skip calibration).


````python
quant_cfg = get_quant_cfg(PRECISION)

if RESUME:
    mo_path = RESUME_MOSTATE or RESUME.replace(".pth", "_mostate.pt")
    print(f"[Resume] Restoring modelopt quantization state from {mo_path}")
    restore_modelopt_state(model, mo_path)
    model = model.to(device)
    ptq_top1 = 0.0
else:
    print("[FP32] Evaluating baseline ...")
    _, fp32_top1, fp32_top5 = validate(model, val_loader, criterion, device)
    print(f"[FP32] Top-1: {fp32_top1:.2f}%  Top-5: {fp32_top5:.2f}%")

    print(f"[PTQ] Calibrating for {PRECISION.upper()} ...")
    model = quantize_model(model, quant_cfg, calib_loader, NUM_CALIB_BATCHES, device)

    _, ptq_top1, ptq_top5 = validate(model, val_loader, criterion, device)
    print(f"[PTQ] Top-1: {ptq_top1:.2f}%  Top-5: {ptq_top5:.2f}%")
````

## Optimizer / Scheduler

Built **after** the quantization graph exists so all quantizer parameters are tracked.


````python
optimizer = torch.optim.SGD(
    model.parameters(), lr=LR, momentum=MOMENTUM, weight_decay=WEIGHT_DECAY
)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS, eta_min=0)
scaler = GradScaler()

start_epoch = 1
best_acc = ptq_top1
if RESUME:
    start_epoch, best_acc = load_training_state(
        RESUME, model, optimizer, scaler, scheduler
    )
    print(f"[Resume] Starting from epoch {start_epoch}, best_acc={best_acc:.2f}%")
````

## QAT fine-tuning loop

````python
print(f"[QAT] Fine-tuning for {EPOCHS} epochs ...")
for epoch in range(start_epoch, EPOCHS + 1):
    print(f"\n[INFO] Epoch {epoch}/{EPOCHS}  lr={optimizer.param_groups[0]['lr']:.6f}")
    train_loss, train_acc = train_one_epoch(
        model, train_loader, criterion, optimizer, scaler, device, epoch
    )
    scheduler.step()
    val_loss, top1, top5 = validate(model, val_loader, criterion, device)
    print(f"  Train loss: {train_loss:.4f}  acc: {train_acc:.2f}%")
    print(f"  Val   loss: {val_loss:.4f}  top-1: {top1:.2f}%  top-5: {top5:.2f}%")

    is_best  = top1 > best_acc
    best_acc = max(top1, best_acc)

    save_checkpoint(
        model=model,
        state={
            "epoch":     epoch,
            "model":     model.state_dict(),
            "optimizer": optimizer.state_dict(),
            "scaler":    scaler.state_dict(),
            "scheduler": scheduler.state_dict(),
            "best_acc":  best_acc,
        },
        directory=OUTPUT_DIR,
        epoch=epoch,
        is_best=is_best,
    )

print(f"\nQAT COMPLETE  --  Best val top-1: {best_acc:.2f}%")
print(f"Best weights: {os.path.join(OUTPUT_DIR, 'qat_modelopt_best.pth')}")
````
